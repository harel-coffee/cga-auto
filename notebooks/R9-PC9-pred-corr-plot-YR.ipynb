```python
# This script generate a correlation plot between BROAD prediction on PC9 and PC9 measurement
# Author: Yiyun

import pandas as pd
from os.path import join
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
anlyze_dir = '/storage/home/yur97/work/cnp_dev/out/21.0423 Lx/L200only_reg_rf_boruta/anlyz'
df_pred = pickle.load(open(join(anlyze_dir,'y_compr_ext.pkl'),'rb'))
```

```python
df_exp_pred = pd.concat([df_pred['actual'],df_pred['predicted']],ignore_index=True).T
df_exp_pred.columns=['actual','predicted']
```

```python
# Plot scatter plot
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
ax = sns.regplot(x='actual', y='predicted',data = df_exp_pred)
ax.set_title('Predicted V.S. Measured PC9 Screening Score', fontsize = 15)
ax.set_xlabel('Measured', fontsize=15);ax.set_ylabel('Predicted', fontsize=15)
# plt.savefig('PC9_exp_pred.pdf')
```

```python

# Spearman correlation
df_exp_pred.corr(method='spearman')
```

```python
# Pearson correlation
df_exp_pred.corr(method='pearson')
```
