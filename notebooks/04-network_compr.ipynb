```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('04-network-compr.csv')
df.head()
```

```python
sns.distplot(df.infer)
sns.distplot(df.assoc)
plt.legend(title='list',loc='upper right', labels=['assoc network', 'inference network'])
plt.xlabel('Node degree')
plt.show()
```
