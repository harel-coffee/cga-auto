```python
dict_code = {}
for i in range(0, len(df_ref19q3.T)):
    name,idx = df_ref19q3.columns[i].split(' ')[0:2]
    if name not in dict_code:
        dict_code[name] = idx
```

```python
def get_HGNC_name(gene_name):
    ext = "/xrefs/symbol/homo_sapiens/"+gene_name+ "?"
    r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
    if not r.ok:
      r.raise_for_status()
      sys.exit()
    decoded = r.json()
    idx = decoded[0]['id']

    ext = "/xrefs/id/"+ idx+ "?external_db=HGNC;species=human;object_type=gene"
    r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
    if not r.ok:
      r.raise_for_status()
      sys.exit()
    decoded = r.json()
    print(len(decoded))
    
    return decoded[0]['display_id']
```
